```python
import os

folder_path = ["./PersianPlagdet2016-text-alignment-train-corpus/src/", "./PersianPlagdet2016-text-alignment-train-corpus/susp/"]

txt_files = []

for folder in folder_path:
    for root, dirs, files in os.walk(folder):
        for file in files:
            if file.endswith(".txt"):
                txt_files.append(os.path.join(root, file)) 
```

```python
len(txt_files)
```

```python
import io


txt_collection = {}

for file in txt_files:
    with io.open(file, 'r', encoding='utf-8') as f:
        txt_collection[file] = f.read()
```

```python
txt_collection[txt_files[2]]
```

```python
document = txt_collection[txt_files[2]]
```

```python
from hazm import *
import tqdm


hazm_normalizer = Normalizer()

txt_sentences = []
txt_norm_tokenize = []


normalized_txt = hazm_normalizer.normalize(document)

sentences = sent_tokenize(normalized_txt)
```

```python
sentences
```

```python
sent_tokenized = []

for sent in sentences:
    sent_tokenized += [word_tokenize(sent)]
```

```python
sent_tokenized
```

```python
import codecs

stop_path = 'persian_stopwords2.txt'
stopwords = [hazm_normalizer.normalize(x.strip()) for x in codecs.open(stop_path,'r','utf-8').readlines()]
```

```python
all_tokens_nonstop = []


for token_list in sent_tokenized:
    temp = []
    temp += [t for t in token_list if t not in stopwords]
    all_tokens_nonstop += [temp]
```

```python
all_tokens_nonstop
```

```python
whole_documnet_nonstop_tokens = []

for sent in all_tokens_nonstop:
    for token in sent:
        whole_documnet_nonstop_tokens += [token]
```

```python
from nltk import FreqDist
import itertools


token_freq_in_doc = FreqDist(list(itertools.chain(itertools.chain(whole_documnet_nonstop_tokens))))
```

```python
token_freq_in_doc
```

```python
from collections import defaultdict
import operator


word_frequencies = defaultdict(int)


for word, frequency in token_freq_in_doc.items():
    word_frequencies[word] += frequency

sorted_frequencies = sorted(word_frequencies.items(), key=operator.itemgetter(1), reverse=True)

print(sorted_frequencies[:5])

most_freq = sorted_frequencies[1][1]
```

```python
most_freq
```

```python
import math


def calc_V(most_freq, word_freq_in_doc, word_freq_in_sent):
    return math.log2(most_freq / (word_freq_in_doc - word_freq_in_sent + 1))
```

```python
sent_freq_ratio = []

punc = ['!', '"', "'", '#', '(', ')', '*', ',', '-', '.', '/', ':', '[', ']', '«', '»', '،', ';', '؛', '@', '$', '%', '^', '&', '+', '_', '=', '?', '؟']



for sent in all_tokens_nonstop:
    token_freq_in_sent = FreqDist(list(itertools.chain(itertools.chain(sent))))

    temp = 0
    counter = 0

    for token in sent:
        if token not in punc:
            temp += calc_V(most_freq, token_freq_in_doc[token], token_freq_in_sent[token])
            print(token, token_freq_in_doc[token], token_freq_in_sent[token])
            counter += 1
    
    sent_freq_ratio += [temp / counter]
```

```python
tagger = POSTagger(model='postagger.model')
```

```python
tagger.tag(word_tokenize('سهراب را دیدم و به او گفتم:(بیا با من برویم!) ]1['))
```

```python
puncs_count = []
pronouns_count = []
word_count = []


for sent in sentences:
    w_count = 0
    p_count = 0
    pro_count = 0

    tags = tagger.tag(word_tokenize(sent))
    for tag in tags:
        if tag[1] == 'PRO':
            pro_count += 1
        
        if tag[1] == 'PUNC':
            p_count += 1
            continue
        
        w_count += 1

    puncs_count += [p_count]
    pronouns_count += [pro_count]
    word_count += [w_count]
```

```python
data = []

for segment in range(len(sentences)):
    data += [[sent_freq_ratio[segment], puncs_count[segment], pronouns_count[segment], word_count[segment]]]
```

```python
data
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Convert the data to a NumPy array
data_array = np.array(data)

# Perform feature scaling
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data_array)

# Instantiate the DBSCAN clustering algorithm
dbscan = DBSCAN(eps=1, min_samples=3)

# Fit the algorithm on the scaled data
dbscan.fit(scaled_data)

# Get the cluster labels
labels = dbscan.labels_

# Apply PCA for dimensionality reduction to 2 dimensions
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(scaled_data)

# Visualize the clusters
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=labels)
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.title('DBSCAN Clustering (PCA)')
plt.show()
```
